```python
import numpy as np
import pandas as pd
```

### MERGING 2 DATAFRAME

```python
students_df = pd.DataFrame({
    'student_id': [101, 102, 103, 104],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [20, 21, 19, 22]
})

grades_df = pd.DataFrame({
    'student_id': [101, 102, 104, 105],
    'subject': ['Math', 'Science', 'History', 'Math'],
    'grade': ['A', 'B', 'A+', 'C']
})
```

```python
students_df
```

```python
grades_df
```

```python
pd.merge(students_df,grades_df)
```

```python
pd.merge(students_df,grades_df,on='student_id',how='inner')
```

```python
pd.merge(students_df,grades_df,on='student_id',how='outer')
```

```python
pd.merge(students_df,grades_df,on='student_id',how='left')
```

```python
pd.merge(students_df,grades_df,on='student_id',how='right')
```
