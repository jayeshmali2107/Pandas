## concatination 2 DATA FRAMES

```python
employees_df = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'dept_id': [101, 102, 103, 104]
})
departments_df = pd.DataFrame({
    'dept_id': [101, 102, 105],
    'department_name': ['HR', 'Engineering', 'Finance']
})
```

```python
employees_df
```

```python
departments_df
```

```python
pd.concat([employees_df,departments_df])
```

```python
pd.concat([departments_df,employees_df])
```

```python
pd.concat([employees_df,departments_df],axis=1)
```
