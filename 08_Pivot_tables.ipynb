```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(1)
dates = np.random.choice(pd.date_range('2025-07-01', '2025-07-05'), size=5)

data = {
    'Date': dates,
    'Product': ['Laptop', 'Phone', 'Monitor', 'Mouse', 'Keyboard'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories'],
    'Units Sold': [3, 5, 2, 10, 6],
    'Unit Price': [60000, 25000, 15000, 1200, 1500]
}

df = pd.DataFrame(data)
df['Total Revenue'] = df['Units Sold'] * df['Unit Price']

df
```

### PIVOT TABLES
it's help for perform aggeration function on tables and select specific part which we needed

```python
pivot_table1 = pd.pivot_table(df,values='Units Sold',index='Category',columns='Product')
pivot_table1
```

```python
pivot_table2 = pd.pivot_table(df,values='Units Sold',index='Product',columns='Category',aggfunc='mean')
pivot_table2
```

```python
pivot_table2 = pd.pivot_table(df,values=['Unit Price','Total Revenue'],index='Product',columns='Category',aggfunc='sum')
pivot_table2
```

### CROSS TABS
it's help for count

```python
cross_table1 = pd.crosstab(df['Category'],df['Product'])
cross_table1
```

```python
cross_table = pd.crosstab(df['Category'],df['Product'])
cross_table
```
